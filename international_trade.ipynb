```python
international_trade_data = pd.read_csv(r'./data/international_trade.csv')
international_trade_data['품목명'].value_counts()
```

```python
# Mapping between item codes and item names in Korean
item_mapping = {
    'TG': '감귤',
    'BC': '꽃양배추와 브로콜리(broccoli)',
    'RD': '무',
    'CR': '당근',
    'CB': '양배추'
}

# Create a reverse mapping from item names to item codes
reverse_item_mapping = {v: k for k, v in item_mapping.items()}

# Replace item names in international_trade_data with item codes
international_trade_data['품목명'] = international_trade_data['품목명'].map(reverse_item_mapping)
international_trade_data = international_trade_data.dropna(subset=['품목명'])  # Drop rows with unmatched item names

international_trade_data
```
